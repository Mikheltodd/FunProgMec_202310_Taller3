Usted es un estudiante de primer semestre que ingresa a la Universidad y quiere estimar su gasto total mensual a partir de su costo de vida como estudiante, información socioeconómica (estrato), lugar de residencia, edad, promedio académico ponderado proyectado en el semestre y el apoyo económico que le ofrece un mecenas  llamado Chocolove McDonell. 

### Componentes del Costo de Vida:

- **Transporte:** el costo del transporte incluye las movilizaciones diarias desde su hogar hacia la Universidad ida y regreso. La tarifa por trayecto es de `$2.800 COP` en el casco urbano de Montería, y de `$3.500 COP` si vive fuera de Montería. 
- **Alimentación:** el día de alimentación se estima en `$20.000 COP` (incluye desayuno, almuerzo y cena). 
- **Matrícula:** el pago de la matrícula se realiza una sola vez por semestre antes de iniciar clases y tiene un valor de `$5’000.000 COP`. 

### Apoyos de Chocolove:

- **Transporte:** Chocolove considera los siguientes auxilios en función de su estrato socioeconómico y edad:
    - Estrato 1: `$65.000 COP` mensuales si es menor de edad; `$50.000 COP` mensuales si es mayor de edad.
    - Estrato 2: `$45.000 COP` mensuales si es menor de edad; `$30.000 COP` mensuales si es mayor de edad.
    - Estrato 3: `$35.000 COP` mensuales si es menor de edad; `$20.000 COP` mensuales si es mayor de edad.
    Los demás estratos no reciben apoyo de transporte de Chocolove. 
- **Alimentación:** Chocolove considera los siguientes auxilios en función de su estrato socioeconómico y edad:
    - Estrato 1: `90%` del costo de alimentación mensual si es menor de edad; `70%` del costo si es mayor de edad.
    - Estrato 2: `70%` del costo de alimentación mensual si es menor de edad; `50%` del costo si es mayor de edad.
    - Estrato 3: `50%` del costo de alimentación mensual si es menor de edad; `30%` del costo si es mayor de edad.
    Los demás estratos no reciben apoyo de alimentación de Chocolove. 
- **Matrícula:** Chocolove regresará hasta el `100%` de la matrícula a los estudiantes que obtengan un promedio ponderado superior a `4.8/5.0` al finalizar el semestre; `70%` si el promedio está entre `4.5` y `4.8`; y `40%` si el promedio se encuentra entre `4.0` y `4.5`.

### Consideraciones:

- Un mes tiene 20 días hábiles.
- El semestre académico corresponde a 4 meses.
- Una persona es mayor de edad si tiene más de 18 años. 

Utiliza los comandos para solicitar información (input()) e imprimir los resultados (print()) a partir de tu información personal.

```python
# Entradas

# edad = 31
# estrato = 4
# promedio = 5
# monteria = True
# dias_mes = 20

edad = int(input("Indique su edad (valor numérico): "))
estrato = int(input("Indique el estrato del lugar donde vive (1 - 6): "))
promedio = float(input("Indique su promedio académico ponderado proyectado (0.0 - 5.0): "))
residencia = input("Vive en Montería? (Y/N): ")
monteria = True if residencia == 'Y' else False
dias_mes = 20

# Costo de Vida:

# Costo Transporte
# Tarifa por trayecto en el casco urbano de Montería
transp_tray_mont = 2800 
# Tarifa por trayecto fuera del casco urbano de Montería
transp_tray_no_mont = 3500

# Uso de Ternary Operator para cálculo del costo de transporte mensual. 
# Se utiliza la variable Booleana 'monteria' que es True si vive en el casco urbano y False de lo contrario.
# Tenga en cuenta que son dos trayectos por día (ida y regreso)
c_transporte_mes = transp_tray_mont * 2 * dias_mes if monteria else transp_tray_no_mont * 2 * dias_mes


# Costo Alimentación
# Alimentación por día
alim_dia = 20000
c_alimentacion_mes = alim_dia * dias_mes

# Costo Matrícula
# Valor de la matrícula por semestre
matricula = 5000000
c_matricula_mes = matricula / 4

# Costo Total
c_total_mes = c_transporte_mes + c_alimentacion_mes + c_matricula_mes

# Apoyos de Chocolove:

# Subsidio Transporte: 

subs_transp_1_menor = 65000
subs_transp_1_mayor = 50000
subs_transp_2_menor = 45000
subs_transp_2_mayor = 30000
subs_transp_3_menor = 35000
subs_transp_3_mayor = 20000

if edad < 18:
    if estrato == 1:
        subs_transporte_mes = subs_transp_1_menor
    elif estrato == 2:
        subs_transporte_mes = subs_transp_2_menor
    elif estrato == 3:
        subs_transporte_mes = subs_transp_3_menor
    else:
        subs_transporte_mes = 0
else:
    if estrato == 1:
        subs_transporte_mes = subs_transp_1_mayor
    elif estrato == 2:
        subs_transporte_mes = subs_transp_2_mayor
    elif estrato == 3:
        subs_transporte_mes = subs_transp_3_mayor
    else:
        subs_transporte_mes = 0
    

# Subsidio Alimentación

p_subs_alim_1_menor = 0.9
p_subs_alim_1_mayor = 0.7
p_subs_alim_2_menor = 0.7
p_subs_alim_2_mayor = 0.5
p_subs_alim_3_menor = 0.5
p_subs_alim_3_mayor = 0.3

if edad < 18:
    if estrato == 1:
        subs_alimentacion_mes = c_alimentacion_mes * p_subs_alim_1_menor
    elif estrato == 2:
        subs_alimentacion_mes = c_alimentacion_mes * p_subs_alim_2_menor
    elif estrato == 3:
        subs_alimentacion_mes = c_alimentacion_mes * p_subs_alim_3_menor
    else:
        subs_alimentacion_mes = 0
else:
    if estrato == 1:
        subs_alimentacion_mes = c_alimentacion_mes * p_subs_alim_1_mayor
    elif estrato == 2:
        subs_alimentacion_mes = c_alimentacion_mes * p_subs_alim_2_mayor
    elif estrato == 3:
        subs_alimentacion_mes = c_alimentacion_mes * p_subs_alim_3_mayor
    else:
        subs_alimentacion_mes = 0

# Subsidio Matrícula
# Entre 4.8 y 5.0
p_subs_exc = 1
# Entre 4.5 y 4.8
p_subs_sobr = 0.7
# Entre 4.0 y 4.5
p_subs_bueno = 0.4

if promedio >= 4.8:
    subs_matricula_mes = c_matricula_mes * p_subs_exc
elif (promedio < 4.8) and (promedio >= 4.5):
    subs_matricula_mes = c_matricula_mes * p_subs_sobr
elif (promedio < 4.5) and (promedio >= 4.0):
    subs_matricula_mes = c_matricula_mes * p_subs_bueno
else:
    subs_matricula_mes = 0

# Subsidio Total
subs_total_mes = subs_transporte_mes + subs_alimentacion_mes + subs_matricula_mes

# Gasto Total
g_total_mes = c_total_mes - subs_total_mes

print("-"*10)
print("Costo de Transporte Mensual: $", c_transporte_mes, "COP")
print("Costo de Alimentación Mensual: $", c_alimentacion_mes, "COP")
print("Costo de Matrícula Mensual: $", c_matricula_mes, "COP")
print("Costo de Vida Mensual: $", c_total_mes, "COP")
print("-"*10)
print("Subsidio de Transporte Mensual: $", subs_transporte_mes, "COP")
print("Subsidio de Alimentación Mensual: $", subs_alimentacion_mes, "COP")
print("Subsidio de Matrícula Mensual: $", subs_matricula_mes, "COP")
print("Subsidio Total Mensual: $", subs_total_mes, "COP")
print("-"*10)
print("Gasto Total Mensual: $", g_total_mes, "COP")
```
